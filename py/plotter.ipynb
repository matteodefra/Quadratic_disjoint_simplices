```python
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset

matplotlib.rcParams['figure.dpi'] = 300
matplotlib.rcParams['figure.figsize'] = 14,8



custom_conf = {'figure.facecolor': 'white',
 'axes.labelcolor': '.15',
 'xtick.direction': 'out',
 'ytick.direction': 'out',
 'xtick.color': '.15',
 'ytick.color': '.15',
 'axes.axisbelow': True,
 'grid.linestyle': '-',
 'text.color': '.15',
 'font.family': ['sans-serif'],
 'font.sans-serif': ['Arial',
  'DejaVu Sans',
  'Liberation Sans',
  'Bitstream Vera Sans',
  'sans-serif'],
 'lines.solid_capstyle': 'round',
 'patch.edgecolor': 'w',
 'patch.force_edgecolor': True,
#  'image.cmap': 'rocket',
 'xtick.top': False,
#  'ytick.right': False,
 'axes.grid': True,
#  'axes.facecolor': '#EAEAF2',
#  'axes.edgecolor': 'white',
 'grid.color': 'lightgrey',
 'axes.spines.left': True,
 'axes.spines.bottom': True,
 'axes.spines.right': True,
 'axes.spines.top': True
#  'xtick.bottom': False,
#  'ytick.left': False
}

# sns.set_theme(style='ticks',rc = {'axes.grid': True})
sns.set_theme(style="ticks", rc = custom_conf)
sns.set_palette('pastel')

palette = sns.color_palette('pastel', 10)

```

```python
'''
 n=1000, K=100
'''

for i in range(0,5):

    if i == 0:
        df0 = pd.read_csv("../results/rule1/n1000/K100/results_n=1000_K=100_update=1_h=1000.0_step=0.csv")
        df1 = pd.read_csv("../results/rule2/n1000/K100/results_n=1000_K=100_update=2_delta=1e-10_h=1e4_step=0.csv")
        df2 = pd.read_csv("../results/rule3/n1000/K100/results_n=1000_K=100_update=3_delta=1e-16_h=1e3_step=0.csv")
    elif i == 1:
        df0 = pd.read_csv("../results/rule1/n1000/K100/results_n=1000_K=100_update=1_h=10000.0_step=1.csv")
        df1 = pd.read_csv("../results/rule2/n1000/K100/results_n=1000_K=100_update=2_delta=1e-16_h=1e6_step=1.csv")
        df2 = pd.read_csv("../results/rule3/n1000/K100/results_n=1000_K=100_update=3_delta=1e-16_h=1e5_step=1.csv")
    elif i == 2:
        df0 = pd.read_csv("../results/rule1/n1000/K100/results_n=1000_K=100_update=1_alpha=1000.0_step=2.csv")
        df1 = pd.read_csv("../results/rule2/n1000/K100/results_n=1000_K=100_update=2_delta=1e-16_alpha=1e6_step=2.csv")
        df2 = pd.read_csv("../results/rule3/n1000/K100/results_n=1000_K=100_update=3_delta=1e-16_alpha=5e2_step=2.csv")
    elif i == 3:
        df0 = pd.read_csv("../results/rule1/n1000/K100/results_n=1000_K=100_update=1_alpha=1000.0_step=3.csv")
        df1 = pd.read_csv("../results/rule2/n1000/K100/results_n=1000_K=100_update=2_delta=1e-16_alpha=1e5_step=3.csv")
        df2 = pd.read_csv("../results/rule3/n1000/K100/results_n=1000_K=100_update=3_delta=1e-16_alpha=5e2_step=3.csv")
    else:
        df0 = pd.read_csv("../results/rule1/n1000/K100/results_n=1000_K=100_update=1_poliak_step=4.csv")
        df1 = pd.read_csv("../results/rule2/n1000/K100/results_n=1000_K=100_update=2_delta=1e4_step=4.csv")
        df2 = pd.read_csv("../results/rule3/n1000/K100/results_n=1000_K=100_update=3_delta=1e-16_step=4.csv")


    # axes = 
    sns.lineplot(
        x=df0['Iteration'],
        y=df0['Dual_gap'],
        alpha=0.7, 
        legend='full',
        lw=2,
        color= palette[0]
    )

    sns.lineplot(
        x=df1['Iteration'],
        y=df1['Dual_gap'],
        alpha=0.7, 
        lw=2,
        color= palette[1]
    )

    sns.lineplot(
        x=df2['Iteration'],
        y=df2['Dual_gap'],
        alpha=0.7, 
        lw=2,
        color= palette[2]
    )

    plt.xscale('log')
    plt.yscale('log')
    plt.ylabel(r"""$f(x^*)$ - $\phi (\lambda)$""", fontsize=18)

    plt.xlabel("Iteration", fontsize=18)

    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    if i == 0:
        plt.legend( title='Steps', 
                    loc='lower left', 
                    labels=[r"""$h=1\mathrm{e}{3}$ (rule 1)""", 
                            r"""$h=1\mathrm{e}{4}$, $\delta=1\mathrm{e}{-10}$ (rule 2)""", 
                            r"""$h=1\mathrm{e}{3}$, $\delta=1\mathrm{e}{-16}$ (rule 3)""",
                            ],
                    fontsize = 'large')
    elif i == 1:
        plt.legend( title='Steps', 
                    loc='lower left', 
                    labels=[r"""$h=1\mathrm{e}{4}$ (rule 1)""", 
                            r"""$h=1\mathrm{e}{6}$, $\delta=1\mathrm{e}{-16}$ (rule 2)""", 
                            r"""$h=1\mathrm{e}{5}$, $\delta=1\mathrm{e}{-16}$ (rule 3)""",
                            ],
                    fontsize = 'large')
    elif i == 2:
        plt.legend( title='Steps', 
                    loc='upper right', 
                    labels=[r"""$\alpha=1\mathrm{e}{3}$ (rule 1)""", 
                            r"""$\alpha=1\mathrm{e}{6}$, $\delta=1\mathrm{e}{-16}$ (rule 2)""", 
                            r"""$\alpha=5\mathrm{e}{2}$, $\delta=1\mathrm{e}{-16}$ (rule 3)""",
                            ],
                    fontsize = 'large')
    elif i == 3:
        plt.legend( title='Steps', 
                    loc='upper right', 
                    labels=[r"""$\alpha=1\mathrm{e}{3}$ (rule 1)""", 
                            r"""$\alpha=1\mathrm{e}{5}$, $\delta=1\mathrm{e}{-16}$ (rule 2)""", 
                            r"""$\alpha=5\mathrm{e}{2}$, $\delta=1\mathrm{e}{-16}$ (rule 3)""",
                            ],
                    fontsize = 'large')
    else:
        plt.legend( title='Steps', 
                    loc='lower left', 
                    labels=[r"""$Poliak$ (rule 1)""", 
                            r"""$Poliak$, $\delta=1\mathrm{e}{4}$ (rule 2)""", 
                            r"""$Poliak$, $\delta=1\mathrm{e}{-16}$ (rule 3)""",
                            ],
                    fontsize = 'large')

    plt.show()
```
